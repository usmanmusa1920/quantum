```python
import qiskit as q

# quantum register
qr = q.QuantumRegister(2)

# classical register
cr = q.ClassicalRegister(2)

# 2 qubits, 2 classical bits
circuit = q.QuantumCircuit(qr, cr)

# not gate, flips qubit 0.
circuit.x(0)

# cnot, controlled not, Flips 2nd qubit's value if first qubit is 1
# (Add a CX (CNOT) gate on control qubit 0 and target qubit 1)
circuit.cx(0, 1)

# ([qubitregister], [classicalbitregister]) Measure qubit 0 and 1 to classical bits 0 and 1
circuit.measure([0,1], [0,1])

# Draw the circuit (Quantum circuit of 2 qubits, 2 classical bits), using matplotlib-based visualization
circuit.draw(output="mpl")
```
